TEST AND TRAIN

```python
x=df.iloc[:,2:32]
x
```

```python
y=df.iloc[:, 1]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x,y,test_size=0.2) #20% test data 
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
```

```python
xtrain = st.fit_transform(xtrain)
xtest= st.fit_transform(xtest)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

MODEL

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(xtrain,ytrain)
```

```python
log.score(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
acc = accuracy_score(ytest,log.predict(xtest))
print("accuracy of your model is :",acc*100,"%")
```

```python
print(classification_report(ytest,log.predict(xtest)))
```

RANDOM FOREST CLASSIFIER 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
acc1 = accuracy_score(ytest,model.predict(xtest))
print("accuracy of your model is :",acc1*100,"%")
```

```python
print(classification_report(ytest,model.predict(xtest)))
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
import math 
math.sqrt(len(ytest))
```

```python
classifier = KNeighborsClassifier(n_neighbors=10, p=2 ,metric='euclidean')
```

```python
classifier.fit(xtrain,ytrain)
```

```python
ypredict = classifier.predict(xtest)
ypredict
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypredict)
print(cm)
```

```python
print(f1_score(ytest,ypredict))
```

```python
acc2=accuracy_score(ypredict,ytest)
print("The accuracy score of your model is : ",acc2*100,"%")
```

```python
print(classification_report(ytest,classifier.predict(xtest)))
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
drt = DecisionTreeClassifier()
drt.fit(xtrain,ytrain)
```

```python
predection1 = drt.predict(xtest)
predection1
```

```python
print("Training score :",drt.score(xtrain, ytrain))
print("Testing score :",drt.score(xtest, ytest))
```

```python
cm1= confusion_matrix(ytest,predection1)
print(cm1)
```

```python
from sklearn.tree import plot_tree
```

```python
lm = DecisionTreeClassifier(max_depth=8)
lm.fit(xtrain,ytrain)
plt.figure("Decision tree",figsize=[50,50])
plot_tree(lm,fontsize=10,filled=True)
plt.show()
```

```python
acc3=accuracy_score(predection1,ytest)
print("The accuracy score of your model is : ",acc3*100,"%")
```

```python
print(classification_report(ytest,drt.predict(xtest)))
```
