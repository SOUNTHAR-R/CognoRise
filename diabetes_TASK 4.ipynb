DIABETES PREDICTION

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import scatter_matrix
import warnings 
warnings.filterwarnings("ignore") 
%matplotlib inline
```

```python
df = pd.read_csv('diabetes_prediction_dataset.csv')
```

```python
df.head(7)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
df["gender"] = label.fit_transform(df["gender"])
df["smoking_history"] = label.fit_transform(df["smoking_history"])
```

```python
df.head(8)
```

```python
df.info()
```

```python
df.isnull().sum()
```

GRAPHICAL REPRESENTATION

```python
df.hist(figsize=(20,10))
plt.show
```

```python
p = scatter_matrix(df, figsize=(20,20))
```

```python
plt.figure(figsize=(12,10))
p=sns.heatmap(df.corr(), annot = True, cmap = "RdYlGn")
```

```python
df.corr()
```

TEST AND TRAIN

```python
from sklearn.model_selection import train_test_split
```

```python
x= df.iloc[:,df.columns != 'diabetes' ]
y= df.iloc[:,df.columns == 'diabetes' ]
```

```python
print(x)
```

```python
print(y)
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x,y,test_size=0.2) #20% test data 
```

```python
xtrain.head()
```

```python
ytrain.head()
```

MODEL 

1.RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
model = RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain.values.ravel())
```

```python
predict_output = model.predict(xtest)
print(predict_output)
```

```python
acc=accuracy_score(predict_output,ytest)
print("The accuracy score of your model is : ",acc*100,"%")
```

```python
print(classification_report(ytest,predict_output))
```

2.LOGISTIC REGRESSION 

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(xtrain,ytrain.values.ravel())
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
prediction = logreg.predict(xtest)
print(prediction)
```

```python
acc1=accuracy_score(prediction,ytest)
print("The accuracy score of your model is : ",acc1*100,"%")
```

```python
print(classification_report(ytest,prediction))
```

K-Nearest Neighbor (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
import math 
math.sqrt(len(ytest))
```

```python
classifier = KNeighborsClassifier(n_neighbors=141, p=2 ,metric='euclidean')
```

```python
classifier.fit(xtrain,ytrain)
```

```python
ypredict = classifier.predict(xtest)
ypredict
```

```python
cm = confusion_matrix(ytest,ypredict)
print(cm)
```

```python
print(f1_score(ytest,ypredict))
```

```python
acc2=accuracy_score(ypredict,ytest)
print("The accuracy score of your model is : ",acc2*100,"%")
```

```python
print(classification_report(ytest,ypredict))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
drt = DecisionTreeClassifier()
drt.fit(xtrain,ytrain)
```

```python
predection1 = drt.predict(xtest)
predection1
```

```python
print("Training score :",drt.score(xtrain, ytrain))
print("Testing score :",drt.score(xtest, ytest))
```

```python
cm1= confusion_matrix(ytest,predection1)
print(cm1)
```

```python
from sklearn.tree import plot_tree
```

```python
lm = DecisionTreeClassifier(max_depth=8)
lm.fit(xtrain,ytrain)
plt.figure("Decision tree",figsize=[50,50])
plot_tree(lm,fontsize=10,filled=True)
plt.show()
```

```python
acc3=accuracy_score(predection1,ytest)
print("The accuracy score of your model is : ",acc3*100,"%")
```

```python
print(classification_report(ytest,predection1))
```

ADABOOSTING CLASSIFIER 

```python
from sklearn.ensemble import AdaBoostClassifier
import warnings 
warnings.filterwarnings("ignore") 
```

```python
adamodel = AdaBoostClassifier(n_estimators=100,learning_rate=1)
```

```python
mo = adamodel.fit(xtrain,ytrain)
dpredict = mo.predict(xtest)
print(dpredict)
```

```python
acc4=accuracy_score(dpredict,ytest)
print("The accuracy score of your model is : ",acc4*100,"%")
```

```python
print(classification_report(ytest,dpredict))
```
